```r
.libPaths(c('/home/richet/Sync/Open/libKriging/alien/libKriging/bindings/R/Rlibs', .libPaths()))
library(rlibkriging)

f <- function(x) 1 - 1/2 * (sin(12*x)/(1+x) + 2*cos(7*x)*x^5 + 0.7)
X <- as.matrix(seq(0.05, 0.95, length.out = 10))
y <- f(X)

wk <- WarpKriging(y, X, warping = 'kumaraswamy', kernel = 'gauss', optim = 'BFGS')
print(wk)
```

```r
x <- as.matrix(seq(0, 1, length.out = 101))
p <- wk$predict(x, return_stdev = TRUE)

p_train <- wk$predict(X, return_stdev = TRUE)
cat('Max interpolation error:', max(abs(p_train$mean - y)), '\n')
cat('RMSE vs true f:', sqrt(mean((p$mean - f(x))^2)), '\n')
```

```r
options(repr.plot.width=8, repr.plot.height=5)
plot(x, f(x), type='l', lty=2, xlab='x', ylab='y',
     main='WarpKriging predict (Kumaraswamy, BFGS)')
polygon(c(x, rev(x)),
        c(p$mean - 2*p$stdev, rev(p$mean + 2*p$stdev)),
        border=NA, col=rgb(0,0,1,0.15))
lines(x, p$mean, col='blue', lwd=2)
lines(x, p$mean - 2*p$stdev, col='blue', lty=2)
lines(x, p$mean + 2*p$stdev, col='blue', lty=2)
points(X, y, col='red', pch=16, cex=1.4)
legend('topright', c('True f','Mean','±2 stdev','Train'),
       lty=c(2,1,2,NA), pch=c(NA,NA,NA,16),
       col=c('black','blue','blue','red'), lwd=c(1,2,1,NA))
```
